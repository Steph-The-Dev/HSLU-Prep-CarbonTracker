```python
import numpy as np
```

```python
winkel_grad = 45
winkel_rad = np.radians(winkel_grad)  # Umrechnung: 45° → π/4
```

```python
cos_theta = np.cos(winkel_rad)
sin_theta = np.sin(winkel_rad)

rotationsmatrix = np.array([
    [cos_theta, -sin_theta],
    [sin_theta,  cos_theta]
])
```

```python
print(rotationsmatrix)
```

```python
vektor = np.array([3, -5])
```

```python
gedrehter_vektor = rotationsmatrix @ vektor

print(f"Original-Vektor: {vektor}")
print(f"Gedrehter Vektor (45°): {gedrehter_vektor}")
```
